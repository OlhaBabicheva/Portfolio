```python
# regular expression matching operations module
import re

# Extract (first) letter!
train['Cabin'] = train['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())

train['Cabin'].unique().tolist()

cabin_category = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'T':8, 'U':9}

train['Cabin'] = train['Cabin'].map(cabin_category)
train['Cabin'].unique().tolist()
#---------------------------------------
test['Cabin'] = test['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())

test['Cabin'].unique().tolist()

cabin_category = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'T':8, 'U':9}

test['Cabin'] = test['Cabin'].map(cabin_category)
test['Cabin'].unique().tolist()
```

```python
title_category = {'Mr':1, 'Miss':2, 'Mrs':3, 'Master':4, 'Other':5}

train['Title'] = train['Title'].map(title_category)
train['Title'].unique().tolist()
```

```python
title_category = {'Mr':1, 'Miss':2, 'Mrs':3, 'Master':4, 'Other':5}

test['Title'] = test['Title'].map(title_category)
test['Title'].unique().tolist()
```

```python
train = train.drop(['Name', 'SibSp', 'Parch', 'Ticket'], axis = 1)
test = test.drop(['Name', 'SibSp', 'Parch', 'Ticket'], axis = 1)
```

```python
train['Survived'] = train['Survived'].astype(np.int8)
```

```python
# module for dataset splitting
from sklearn.model_selection import train_test_split

# local validation scheme
X_train, X_test, y_train, y_test = train_test_split(train.drop(['Survived', 'PassengerId'], axis = 1), 
                                                    train['Survived'], test_size = 0.2, 
                                                    random_state = 2)
```

```python
# logistic regression model 
from sklearn.linear_model import LogisticRegression
logisticRegression = LogisticRegression(max_iter = 10000)
logisticRegression.fit(X_train, y_train)
```

```python
predictions = logisticRegression.predict(X_test)
```

```python
# Check mean
round(np.mean(predictions), 2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# resulting confusion matrix
print(confusion_matrix(y_test, predictions))
```

```python
# Accuracy calculation
accuracy = (87 + 55) / (87 + 13 + 24 + 55)
print('accuracy is: ' + str(round(accuracy, 2)))
```

```python
from sklearn.model_selection import KFold

# robust cross-validation scheme
kf = KFold(n_splits = 5, random_state = 2)
```

```python
from sklearn.model_selection import cross_val_score

# CV accuracy estimate:
print(cross_val_score(logisticRegression, X_test, y_test, cv = kf).mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

#Initialize randomForest
randomForest = RandomForestClassifier(random_state = 2)
```

```python
# Set parameter grid
param_grid = { 
    'criterion' : ['gini', 'entropy'],
    'n_estimators': [100, 300, 500],
    'max_features': ['auto', 'log2'],
    'max_depth' : [3, 5, 7]    
}
```

```python
from sklearn.model_selection import GridSearchCV

# Grid search
randomForest_CV = GridSearchCV(estimator = randomForest, param_grid = param_grid, cv = 5)
randomForest_CV.fit(X_train, y_train)
```

```python
# Print best hyperparameters
randomForest_CV.best_params_
```

```python
# Define optimal randomForest algo
randomForestFinalModel = RandomForestClassifier(random_state = 2, criterion = 'gini', max_depth = 7, max_features = 'auto', n_estimators = 300)
```

```python
# Fit the model to the training set
randomForestFinalModel.fit(X_train, y_train)
```

```python
# Prediction
predictions = randomForestFinalModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# accuracy for random forest
print("accuracy is: ", round(accuracy_score(y_test, predictions), 2))
```

```python
# Prediction
test['Survived'] = randomForestFinalModel.predict(test.drop(['PassengerId'], axis = 1))
```

```python
# test predictions for final submission
test[['PassengerId', 'Survived']].to_csv('rf_submission.csv', index = False)
```
